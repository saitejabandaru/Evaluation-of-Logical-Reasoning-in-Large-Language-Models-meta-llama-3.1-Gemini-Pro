```python
import transformers
import torch

model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

# Update the configuration for rope_scaling *before* loading the pipeline
config = transformers.AutoConfig.from_pretrained(model_id, trust_remote_code=True)
# Directly set the rope scaling within the config dictionary, using 'name' and 'factor'
config.update({"rope_scaling": {"type": "linear", "factor": 8.0}})


pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    config=config,  # Pass the updated config
    model_kwargs={"torch_dtype": torch.bfloat16},  # Pass trust_remote_code to model_kwargs
    device_map="auto"
)
```

```python
# Specify the path to your file
file_path = "/content/LogiQA-dataset/Test.txt"

# Open the file and read its contents as a single string
with open(file_path, "r", encoding="utf-8") as file:
    file_content = file.read()

# Print the content or work with it as a single string
#print(file_content)
```

```python
file_content
```

```python
# Split file content into lines
line_list = file_content.split("\n")
Questions = []

# Initialize variables to store each instance's data
correct_option = ""
passage = ""
question = ""
choices = {}

# Parse each 8-line instance
for i in range(0, len(line_list), 8):
    # Each 8 lines make up one question instance
    if i + 7 < len(line_list):
        blank_line = line_list[i].strip()  # First line is blank
        correct_option = line_list[i + 1].strip()  # Second line is the correct option
        passage = line_list[i + 2].strip()  # Third line is the context/passage
        question = line_list[i + 3].strip()  # Fourth line is the question
        choices = {
            "A": line_list[i + 4].strip()[2:].strip(),  # Fifth to eighth lines are choices A-D
            "B": line_list[i + 5].strip()[2:].strip(),
            "C": line_list[i + 6].strip()[2:].strip(),
            "D": line_list[i + 7].strip()[2:].strip(),
        }

        # Append the parsed instance to Questions list
        Questions.append({
            "passage": passage,
            "question": question,
            "choices": choices,
            "correct_option": correct_option
        })

# Now `Questions` contains the structured data for each question instance
print(f"Parsed {len(Questions)} questions.")
```

# Accuracy

```python
results = []
correct_count = 0
for question in Questions[1:]:
  # Construct a conversational prompt
  messages = [
      {"role": "system", "content": "You are an assistant who provides answers to logical reasoning questions. Use Chain of thought reasoning but dont give Chain of Thought output, Just give the Single Letter correct option"},
      {"role": "user", "content": f"{question['passage']}\nChoices:\n" +
                                  "\n".join([f"{k}: {v}" for k, v in question['choices'].items()]) +
                                  "\nAnswer the correct choice (A, B, C, or D)."}
  ]
  output = pipeline(messages, max_new_tokens=50)
  print(output[0]['generated_text'][-1]['content'])
  correct_answer = question['correct_option']
  predicted_answer = output[0]['generated_text'][-1]['content']
  is_correct = (predicted_answer.lower() == correct_answer.lower())

  results.append({
      'passage': question['passage'],
      'question': question['question'],
      'predicted_answer': predicted_answer,
      'correct_answer': correct_answer,
      'is_correct': is_correct
  })
  if is_correct:
    correct_count += 1

accuracy = correct_count / len(Questions)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Specify the path to your file
file_path = "/content/argument-reasoning-comprehension-task/mturk/annotation-task/data/exported-SemEval2018-train-dev-test/test-full.txt"

# Open the file and read its contents as a single string
with open(file_path, "r", encoding="utf-8") as file:
    file_content = file.read()
```

```python
file_content
```

```python
# Re-attempting the parsing process without including the correct label.
# The focus will be on reason, claim, warrant0, and warrant1 for each ID.

# Dictionary to hold parsed data by ID
parsed_data = {}
correct_labels = {}

try:
    with open(file_path, 'r', encoding='utf-8') as file:
        # Skip the header line
        next(file)

        # Read and parse each line
        for line in file:
            # Split the line by tab
            parts = line.strip().split('\t')

            # Ensure the line has the expected number of columns
            if len(parts) >= 8:
                # Assign values to each variable
                instance_id = parts[0]
                warrant0 = parts[1]
                warrant1 = parts[2]
                reason = parts[4]
                claim = parts[5]
                debate_title = parts[6]
                debate_info = parts[7]

                # Store data in a structured dictionary by instance ID, excluding the correct label
                parsed_data[instance_id] = {
                    "reason": reason,
                    "claim": claim,
                    "warrant0": warrant0,
                    "warrant1": warrant1,
                    "debate_title": debate_title,
                    "debate_info": debate_info
                }
                correct_labels[instance_id] = parts[3]

except FileNotFoundError as e:
    parsed_data = str(e)

# Display a sample of the parsed data (first 5 instances)

```

```python
list(parsed_data.items())[:5]
```

```python
list(correct_labels.items())[:5]
```

```python
# Initialize counters
correct_predictions = 0
total_instances = len(parsed_data)

results = []

# Loop through each instance in the dataset
for instance_id, data in parsed_data.items():
    # Generate the prompt for each question
    prompt = f"""
      Premise: {data['reason']}
      Claim: {data['claim']}

      Which of the following warrants makes the claim logically consistent with the premise?

      Option 0: {data['warrant0']}
      Option 1: {data['warrant1']}

      Please respond with the correct option number (0 or 1).
    """
    messages = [
      {"role": "system", "content": "You are an assistant who provides answers to logical reasoning questions. Use Chain of thought reasoning but dont give Chain of Thought output, Just give the Single Letter correct option, 0 for Option 1 and 1 Option 1."},
      {"role": "user", "content": prompt}
    ]

    try:
        output = pipeline(messages, max_new_tokens=50)
        print(output[0]['generated_text'][-1]['content'])

        predicted_answer = output[0]['generated_text'][-1]['content']
        correct_answer = correct_labels[instance_id]

        # Check if the prediction is correct
        is_correct = (predicted_answer == correct_answer)

        if is_correct:
            correct_predictions += 1

        results.append({
                "instance_id": instance_id,
                "predicted_answer": predicted_answer,
                "correct_answer": correct_answer,
                "is_correct": is_correct
        })

    except Exception as e:
        print(f"Error processing instance {instance_id}: {e}")


accuracy = correct_predictions / total_instances
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
results = []

# Loop through each instance in the dataset
for instance_id, data in parsed_data.items():
    # Generate the prompt for each question
    prompt = f"""
      debate title: {data['debate_title']}
      debate info: {data['debate_info']}
      Premise: {data['reason']}
      Claim: {data['claim']}

      Which of the following warrants makes the claim logically consistent with the premise?

      Option 0: {data['warrant0']}
      Option 1: {data['warrant1']}

      let's think step by step.
    """
    messages = [
      {"role": "system", "content": "You are an assistant who provides answers to logical reasoning questions. Use Chain of thought reasoning to answer the question and give reasining chains leading to the correct answer as as output"},
      {"role": "user", "content": prompt}
    ]

    try:
        output = pipeline(messages, max_new_tokens=50)
        predicted_answer = output[0]['generated_text'][-1]['content']
        print(predicted_answer)

        results.append({
                "instance_id": instance_id,
                "debate_title": {data['debate_title']}
                "debate_info": {data['debate_info']}
                "Premise": {data['reason']}
                "Claim": {data['claim']}
                "predicted_answer": predicted_answer,
                "correct_answer": correct_answer
        })

    except Exception as e:
        print(f"Error processing instance {instance_id}: {e}")
```
